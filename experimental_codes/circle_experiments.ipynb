```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
image = cv2.imread("TestImages4Project//butterfly.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.show()
```

```python
color = (0, 0, 255)
```

```python
radius = 20
```

```python
center_coordinates = (20,20)
```

```python
center_coordinates1 = (100,100)
```

```python
thickness = 2
```

```python
image = cv2.circle(image, center_coordinates, radius, color, thickness)
```

```python
image = cv2.circle(image, center_coordinates1, radius, color, thickness)
```

```python
cv2.imshow('Circle Image', image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
